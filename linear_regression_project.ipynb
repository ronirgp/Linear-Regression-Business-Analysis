# Linear Regression Business Analysis

## 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python

df = pd.read_csv("data/business_data.csv")

```

```python

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.scatter(df["Marketing_Spend"], df["Units"])

plt.xlabel("Marketing Spend")
plt.ylabel("Units Sold")
plt.title("Marketing Spend vs Units Sold")

plt.show()
```

```python
X = df[["Marketing_Spend"]]
y = df["Units"]
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
print("Intercept (b0):", model.intercept_)
print("Slope (b1):", model.coef_[0])
```

```python
predictions = model.predict(X)

print(predictions)
```

```python
comparison = df.copy()

comparison["Predicted_Units"] = predictions

print(comparison)
```

```python
plt.scatter(df["Marketing_Spend"], df["Units"], color="blue", label="Actual Sales")

plt.plot(df["Marketing_Spend"], predictions,
         color="red",
         linewidth=2,
         label="Regression Line")

plt.xlabel("Marketing Spend")
plt.ylabel("Units Sold")
plt.title("Linear Regression Model")
plt.legend()

plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, predictions)

print("R² Score:", r2)
```

```python
comparison = df.copy()

comparison["Predicted_Units"] = predictions
comparison["Error"] = comparison["Units"] - comparison["Predicted_Units"]
comparison["Absolute_Error"] = comparison["Error"].abs()

comparison
```

```python
comparison.to_csv("results/regression_results.csv", index=False)

print("Results saved successfully!")
```

```python
from sklearn.metrics import root_mean_squared_error, r2_score

rmse = root_mean_squared_error(y, predictions)
r2 = r2_score(y, predictions)

print("RMSE:", rmse)
print("R²:", r2)
```

```python
with open("results/model_summary.txt", "w") as file:
    file.write("Linear Regression Business Analysis\n")
    file.write("=" * 40 + "\n\n")

    file.write(f"Intercept (b0): {model.intercept_:.4f}\n")
    file.write(f"Slope (b1): {model.coef_[0]:.4f}\n")
    file.write(f"RMSE: {rmse:.4f}\n")
    file.write(f"R² Score: {r2:.4f}\n\n")

    file.write("Model Description:\n")
    file.write("This model predicts Units Sold based on Marketing Spend.\n")
    file.write("The predictions are estimates and may differ from actual sales.\n")
    file.write("Adding more variables such as weather, holidays, region,\n")
    file.write("product type, and pricing could improve the model.\n")

print("Model summary saved successfully!")
```
